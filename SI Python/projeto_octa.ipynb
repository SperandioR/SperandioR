```python
import matplotlib.pyplot as plt
x = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
y = [2.74, 2.88, 2.98, 3.32, 3.73, 3.68, 4.19, 4.30, 4.46, 5.49, 6.99]
plt.title('Preço Médio da Gasolina no Brasil', fontsize=15)
plt.plot(x, y, c='red', ls='-.', lw='1',marker='o')
plt.ylabel('Preço', fontsize=16)
plt.xlabel('Ano', fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
names= ['Atlético-Mg','Corinthians', 'Coritiba', 'São Paulo', 'Santos', 'Botafogo', 'Flamengo']
vals = [12, 13, 10, 11, 10,11, 6 ]
fig, ax = plt.subplots(figsize = (8, 5))
ax.bar(names, vals, color=['black','Brown', 'Green', 'Salmon', 'y', 'Purple', 'Red'])

ax.set_title('Brasileirão Série A', fontsize = 18)

ax.set_ylabel('Pontuação', fontsize = 14)
ax.set_xlabel('Times', fontsize = 14)

for idx, val in enumerate(vals):
    txt = f'{val} Pts'
    y_coord = val + 0.1
    x_coord = idx - 0.30

    ax.text(x=x_coord, y=y_coord, s=txt, fontsize=12)

ax.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
  
names= ['Corinthians', 'Sujit', 'Sayan', 'Saikat'] 
x=np.arange(4) 
marks=[ 60, 90, 55, 46] 
error=[ 11, 15, 5, 9] 
colors = ['red', 'green', 'blue', 'black'] 
fig, ax = plt.subplots() 
ax.bar(x, marks, alpha = 0.5, 
       color = colors) 
  
for pos, y, err, colors in zip(x, marks, 
                               error, colors): 
    
    ax.errorbar(pos, y, err, lw = 2, 
                capsize = 4, capthick = 4, 
                color = colors) 
      
ax.set_ylabel('Marks of the Students') 
  
ax.set_xticks(x) 
ax.set_xticklabels(names) 
ax.set_xlabel('Name of the students') 
plt.show()
```
