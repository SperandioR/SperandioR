```python
import matplotlib.pyplot as plt
x = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
y = [2.74, 2.88, 2.98, 3.32, 3.73, 3.68, 4.19, 4.30, 4.46, 5.49, 6.99]
plt.title('Preço Médio da Gasolina no Brasil', fontsize=15)
plt.plot(x, y, c='red', ls='-.', lw='1',marker='o')
plt.ylabel('Preço', fontsize=16)
plt.xlabel('Ano', fontsize=16)
plt.show()
```
