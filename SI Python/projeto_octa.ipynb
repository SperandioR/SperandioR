```python
import matplotlib.pyplot as plt
names= ['Atlético-Mg','Corinthians', 'Coritiba', 'São Paulo', 'Santos', 'Botafogo', 'Flamengo', 'Palmeiras', 'Fortaleza']
vals = [12, 14, 10, 12, 11,12,9,12,1]
fig, ax = plt.subplots(figsize = (11, 5))
ax.bar(names, vals, color=['black','Brown', 'Green', 'Salmon', 'y', 'Purple', 'Red','blue','Pink'])

ax.set_title('Brasileirão Série A', fontsize = 18)

ax.set_ylabel('Pontuação', fontsize = 14)
ax.set_xlabel('Times', fontsize = 14)
for idx, val in enumerate(vals):
    txt = f'{val} Pts'
    y_coord = val + 0.1
    x_coord = idx - 0.30

    ax.text(x=x_coord, y=y_coord, s=txt, fontsize=12)

ax.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
  
names= ['Corinthians', 'Sujit', 'Sayan', 'Saikat'] 
x=np.arange(4) 
marks=[ 60, 90, 55, 46] 
error=[ 11, 15, 5, 9] 
colors = ['red', 'green', 'blue', 'black'] 
fig, ax = plt.subplots() 
ax.bar(x, marks, alpha = 0.5, 
       color = colors) 
  
for pos, y, err, colors in zip(x, marks, error, colors): 
    
    ax.errorbar(pos, y, err, lw = 2, 
                capsize = 4, capthick = 4, 
                color = colors) 
      
ax.set_ylabel('Marks of the Students') 
  
ax.set_xticks(x) 
ax.set_xticklabels(names) 
ax.set_xlabel('Name of the students') 
plt.show()
```

```python
# Importando numpy
import numpy as np
import matplotlib.pyplot as plt 
# Dados para compor o gráfico
labels = ['Elon Musk', 'Jeff Bezos', 'Bill Gates','Mark Zuckerberg', 'Larry Page']
# Definindo a quantidade de cada dado
data = [219.0, 171.0, 129.0, 67.3, 111.0] 
# Aqui criamos a área que plotamos o gráfico e definimos seu tamanho
fig, ax = plt.subplots(figsize=(11, 6), subplot_kw=dict(aspect="equal"))
# Criando o gráfico e colocando a função da legenda interna
wedges, texts, autotexts = ax.pie(data, 
    labels=labels, 
    autopct=lambda pct: func(pct, data),
    textprops=dict(color="Black"), 
    explode=(0.05,0.05,0.05,0.05,0.05))
# Aqui serão colocados os kg e as porcentagens no gráfico
def func(pct, allvals):
    # calc %
    absolute = int(pct/100.*np.sum(allvals))
    # fazendo legenda do gráfico com % e kg
    return "{:.1f}%\n({:d} Bi)".format(pct, absolute)
# Aqui definimos o tamanho do texto de dentro do gráfico, e o peso da fonte como bold
plt.setp(autotexts, size=9, weight="bold")
# Título do gráfico
ax.set_title('Forbes Billionaires 2022')
# Mostrando o gráfico
plt.show()
```
