```python
import matplotlib.pyplot as plt
x = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
y = [4.5, 5.0, 5.5, 6.0, 6.5, 7.0, 7.5, 8.0]
plt.title('Preço Médio da Gasolina no Brasil', fontsize=15)
plt.plot(x, y, c='red', ls='-.', lw='1',marker='o')
plt.ylabel('Preço', fontsize=16)
plt.xlabel('Ano', fontsize=16)
plt.show()
```
